## Building Permits dataset from data.calgary.ca

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/building_permits.csv')
```

```python
# run all siz inspections commands
print('Shape:', df.shape)
print('\nFirst 5 rows:')
print(df.head())
print('\nColumn info:')
print(df.info())
print('\nNumeric summary:')
print(df.describe())
```

```python
# checking missing values
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_report = pd.DataFrame({'missing_count': missing, 'missing_percent': missing_percent})
print(missing_report[missing_report['missing_count'] > 0])
```
